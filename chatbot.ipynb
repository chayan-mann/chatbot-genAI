```python
import os
import json
import getpass
import boto3
from botocore.config import Config
from opensearchpy import OpenSearch
from langchain.llms.bedrock import Bedrock
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate


# Initialize Bedrock client
retry_config = Config(
        region_name = 'us-east-1',
        retries = {
            'max_attempts': 10,
            'mode': 'standard'
        }
)
session = boto3.Session(
    region_name = 'us-east-1',
    aws_access_key_id= getpass.getpass("Enter your AWS Access Key ID: "),
    aws_secret_access_key= getpass.getpass("Enter your AWS Secret Access Key: ")
)

bedrock_runtime = session.client(
    service_name = 'bedrock-runtime',
    region_name = 'us-east-1',
    config = retry_config
)

```

```python
from opensearchpy import OpenSearch

CLUSTER_URL = 'https://search-personate-chat-v1-6sehxcwx4sit2ewbsa6psro3yu.us-east-2.es.amazonaws.com'

def get_client(cluster_url = CLUSTER_URL,
                  username='rishab',
                  password= getpass.getpass("Enter your password")):

    client = OpenSearch(
        hosts=[cluster_url],
        http_auth=(username, password),
        verify_certs=False
    )
    return client

client = get_client()
```

```python
import json
from langchain.llms.bedrock import Bedrock

message_list = []

initial_message = {
    "role": "user",
    "content": [
        { "text": input("Ask a question:-") }
    ],
}

message_list.append(initial_message)

response = bedrock_runtime.converse(
    modelId="anthropic.claude-3-sonnet-20240229-v1:0",
    messages=message_list,
    inferenceConfig={
        "maxTokens": 2000,
        "temperature": 0
    },
)

response_message = response['output']['message']
message_list.append(response_message)
print(json.dumps(message_list, indent=4))
```
